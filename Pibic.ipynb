```python
from scipy.stats.stats import pearsonr

col =  ['cloroquina', 'ivermectina',
       'hidroxicloroquina', 'nitazoxanida', 'zinco', 'vitaminaD', 'vitaminaC',
       'azitromicina']
res_r = []
res_p = []
res_res = []

for k in col:
    res_r.append(round(pearsonr(dados['obitos'], dados[k])[0], 4))

    #extract p-value (rounded to 4 decimal places)
    res_p.append(round(pearsonr(dados['obitos'], dados[k])[1], 4))

for i in range(0,8):
    if res_p[i] < 0.05:
        res_res.append('HÃ¡ uma associaÃ§Ã£o')
    else:
        res_res.append('NÃ£o hÃ¡ uma associaÃ§Ã£o')

valores = pd.DataFrame(zip(col, res_r, res_p, res_res), columns = ['Termo de Busca', 'CorrelaÃ§Ã£o', 'p-valor', 'ConclusÃ£o'])

valores.to_csv('obitos_tabela.csv', index = False)




```
